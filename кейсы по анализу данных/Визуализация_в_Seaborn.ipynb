```python
# Импортируем Seaborn:
import seaborn as sns

# Устанавливаем тему по умолчанию:
sns.set_theme()

# Загружаем набор данных для примера:
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# Строим график:
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    col='time',
    hue='sex',
    style='smoker',
    size='size');
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins
```

```python
sns.relplot(data=penguins,
            kind='line',
            x='body_mass_g',
            y='flipper_length_mm',
            col='island',
            row='sex',
            hue='species',
            ci=None,
            facet_kws={'sharex': False});
```

```python
sns.relplot(data=penguins,
            kind='line',
            x='body_mass_g',
            y='flipper_length_mm',
            hue='species',
            style='island',
            size='sex',
            ci=None,);
```

```python
sns.lmplot(data=penguins,
            x='body_mass_g',
            y='flipper_length_mm',
            col='island',
            row='sex',
            hue='species');
```

```python
sns.displot(data=penguins,
            x='body_mass_g',
            hue='sex',
            col='species',
            bins=30,
            kde=True);
```

```python
sns.displot(data=penguins,
            x='body_mass_g',
            y='flipper_length_mm',
            hue='sex',
            col='species',
            kind='kde');
```

```python
sns.catplot(data=penguins,
            x='species',
            y='body_mass_g',
            hue='sex');
```

```python
sns.catplot(data=penguins,
            x='species',
            y='body_mass_g',
            hue='sex',
            kind='box');
```

```python
sns.boxplot(data=penguins,
            x='species',
            y='body_mass_g',
            hue='sex');
```

```python
sns.catplot(data=penguins,
            x='species',
            y='body_mass_g',
            col='island',
            hue='sex',
            kind='box');
```

```python
sns.jointplot(data=penguins,
              x='flipper_length_mm',
              y='body_mass_g',
              hue='species');
```

```python
sns.pairplot(penguins, hue='sex');
```

```python
g = sns.PairGrid(penguins, hue="species")
g.map_upper(sns.histplot, bins=30)
g.map_lower(sns.kdeplot, bw_adjust=0.7)
g.map_diag(sns.histplot, kde=True);
```
