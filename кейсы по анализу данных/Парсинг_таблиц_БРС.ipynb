# Анализ БРС

```python
import pandas as pd
import numpy as np
```

```python
url = "https://rating.unecon.ru/index.php?&y=2019&k=1&f=1&up=12270&g=all&upp=all&sort=fio&ball=hide&s=1"
```

```python
df = pd.read_html(url, header=1, index_col="№", encoding="utf8")
df[0].head()
```

```python
df = df[0]
```

```python
df.describe()
```

```python
df.groupby('№ группы')['∑баллов'].describe()
```

```python
group = df[df['№ группы'] == "Л-1905"]
group 
```

```python
group_newindex = group.set_index(['Фамилия, имя, отчество'])
group_newindex
```

```python
group_newindex.loc["Баринова Анастасия Дмитриевна"][["ДЯиК", "ИЛУ", "ОПДП", "ОЯ", "ПКПИЯ"]].plot();
```

```python
max_val = group["∑баллов"].max()
max_val
```

```python
group[group["∑баллов"] == max_val]
```
