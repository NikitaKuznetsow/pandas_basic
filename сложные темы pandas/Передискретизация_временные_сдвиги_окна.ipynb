```python
import pandas_datareader.data as web
```

```python
# https://pandas-datareader.readthedocs.io/en/latest/readers/moex.html
f = web.DataReader('USD000UTSTOM', 'moex', start='2020-07-15', end='2020-07-31')
f.head()
```

```python
f = f.CLOSE
f[:10]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set()

f.plot();
```

```python
f.plot(alpha=0.5, style='-')
f.resample('BA').mean().plot(style=':')
f.asfreq('BA').plot(style='--');
plt.legend(['input', 'resample', 'asfreq'], loc='upper left');
```
